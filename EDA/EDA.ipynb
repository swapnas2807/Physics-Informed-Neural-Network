# Exploratory Data Analysis (EDA)
This notebook performs basic EDA on the dataset `FINAL_4CLASSES.csv`, including range calculation for each column.

## 1. Import Required Libraries
Import pandas and numpy for data manipulation and analysis.

```python
import pandas as pd
import numpy as np
```

## 2. Load the Dataset
Read the dataset into a pandas DataFrame.

```python
# Update the path if your CSV is in a different location
file_path = '../Data/FINAL_4CLASSES.csv'
df = pd.read_csv(file_path)
df.shape
```

## 4. Check for Missing Values
Check for missing values in each column.

```python
# Missing values per column
df.isnull().sum()
```

## 5. Calculate Ranges for Each Column
For each numeric column, calculate and display the minimum and maximum values.

```python
# Ranges for numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    print(f"{col}: min = {df[col].min()}, max = {df[col].max()}")
```

## 6. Summary Statistics for Numeric Columns
Show count, mean, std, min, max, and quartiles for numeric columns.

```python
# Summary statistics for numeric columns
df.describe()
```

## 7. Unique Values in Categorical Columns
For each categorical column, display the number of unique values and list the unique entries.

```python
# Unique values in categorical columns (based on FINAL_4CLASSES.csv)
cat_cols = ['Class']
for col in cat_cols:
    print(f"{col}: {df[col].nunique()} unique values")
    print(df[col].unique())
    print('-'*40)
```

## Value Distance Between Unique Values in Each Input Column
This cell calculates the difference between each unique value in every input parameter column.

```python
# List of input parameter columns (excluding outputs and class)
input_columns = [
    'Temperature(°)', 'W12(um)', 'W34(um)', 'W58(um)', 'W6(um)', 'W7(um)', 'Idc(uA)', 'Length(um)'
]

for col in input_columns:
    unique_vals = np.sort(df[col].unique())
    if len(unique_vals) > 1:
        diffs = np.diff(unique_vals)
        print(f"{col}: {len(unique_vals)} unique values, value differences: {diffs}")
    else:
        print(f"{col}: Only one unique value: {unique_vals[0]}")
```
